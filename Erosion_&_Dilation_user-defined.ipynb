Lab Question 9:
Perform morphological operations to get the resultant image. Define your function for dilation.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('/content/j.png',0)
_, img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
plt.imshow(img,cmap='gray')
```

```python
# Erosion
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
erosion = cv2.erode(img, kernel, iterations=1)
plt.imshow(erosion,cmap='gray')
plt.title("Eroded Image")
```

```python
# Dilation
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
dilate = cv2.dilate(img, kernel, iterations=1)
plt.imshow(dilate,cmap='gray')
plt.title("Dilated Image")
```

```python
size = 5
kernel = np.ones([size,size])*255

height, width = img.shape
dil_img = np.zeros([height, width])
ero_img = np.zeros([height, width])
```

```python
for i in range(height - size):
    for j in range(width - size):
        if np.all(img[i:i+size, j:j+size] == kernel):
            ero_img[i, j] = 255

plt.imshow(ero_img, cmap='gray')
plt.title("Eroded Image")
```

```python
for i in range(height - size):
    for j in range(width - size):
        if np.any(img[i:i+size, j:j+size] == kernel):
            dil_img[i, j] = 255

plt.imshow(dil_img, cmap='gray')
plt.title("Dilated Image")
```
